Use the data `loans_income.csv` to do the following exercise:

Read the data in pandas and:

1. Draw a sample with 1000 observations. Plot the sample's data distribution
2. Implement bootstrap to take 10, 50, 100, and 1000 times resampling to repeatedly draw an income sample with 10 incomes. Plot the sample mean distribution for each resampling 
3. Implement bootstrap to take 10, 50, 100, and 1000 times resampling to repeatedly draw an income sample with 100 incomes. Plot the sample mean distribution for each resampling 
4. Overlap distribution plots of samples in 2 and 3 when the resampling size is the same.


```python
#1. Draw a sample with 1000 observations. Plot the sample's data distribution
#(Without replacement)
import pandas as pd
LI=pd.read_csv("loans_income.csv")
Sample1=LI.sample(n=1000, random_state=0)
Sample1.hist(bins=200)
```

```python
#2. Implement bootstrap to take 10, 50, 100, and 1000 times resampling to repeatedly draw an income sample with 10 incomes. 
#Plot the sample mean distribution for each resampling 
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

def bstrap(n_resample, samplesize):
    sample_mean = []
    for i in range(n_resample):
      y = random.sample(LI['x'].tolist(), samplesize)
      avg = np.mean(y)
      sample_mean.append(avg)
    return sample_mean

figure, axis = plt.subplots(2, 2)

axis[0,0].hist(bstrap(10,10),bins=10)
axis[0,1].hist(bstrap(50,10),bins=50)
axis[1,0].hist(bstrap(100,10),bins=100)
axis[1,1].hist(bstrap(1000,10),bins=1000)

axis[0,0].set_title("Sample time=10",fontsize=7)
axis[0,1].set_title("Sample time=50",fontsize=7)
axis[1,0].set_title("Sample time=100",fontsize=7)
axis[1,1].set_title("Sample time=1000",fontsize=7)

plt.show()
```

```python
#3. Implement bootstrap to take 10, 50, 100, and 1000 times resampling to repeatedly draw an income sample with 100 incomes. Plot the sample mean distribution for each resampling 


figure, axis = plt.subplots(2, 2)

axis[0,0].hist(bstrap(10,100),bins=10)
axis[0,1].hist(bstrap(50,100),bins=50)
axis[1,0].hist(bstrap(100,100),bins=100)
axis[1,1].hist(bstrap(1000,100),bins=1000)

axis[0,0].set_title("Sample time=10",fontsize=7)
axis[0,1].set_title("Sample time=50",fontsize=7)
axis[1,0].set_title("Sample time=100",fontsize=7)
axis[1,1].set_title("Sample time=1000",fontsize=7)

plt.show()
```

```python
#4. Overlap distribution plots of samples in 2 and 3 when the resampling size is the same.
plt.rc('xtick', labelsize=6) 
plt.rc('ytick', labelsize=6) 
#Plot 1
plt.subplot(2, 2, 1)
plt.hist(bstrap(10,10), bins=100, alpha=0.5, label='samplesize=10')
plt.hist(bstrap(10,100), bins=100, alpha=0.5, label='samplesize=100')
plt.legend(loc='upper right')
plt.title('Resampling Size=10')

#Plot 2
plt.subplot(2, 2, 2)
plt.hist(bstrap(50,10), bins=50, alpha=0.5, label='samplesize=10')
plt.hist(bstrap(50,100), bins=50, alpha=0.5, label='samplesize=100')
plt.legend(loc='upper right')
plt.title('Resampling Size=50')

#Plot 3
plt.subplot(2, 2, 3)
plt.hist(bstrap(100,10), bins=100, alpha=0.5, label='samplesize=10')
plt.hist(bstrap(100,100), bins=100, alpha=0.5, label='samplesize=100')
plt.legend(loc='upper right')
plt.title('Resampling Size=100',fontsize=5,pad=2.0)

#Plot 4
plt.subplot(2, 2, 4)
plt.hist(bstrap(1000,10), bins=100, alpha=0.5, label='samplesize=10')
plt.hist(bstrap(1000,100), bins=100, alpha=0.5, label='samplesize=100')
plt.legend(loc='upper right')
plt.title('Resampling Size=1000',fontsize=5,pad=2.0)
plt.show()
```
