```python
#1. Show all data types, missing rates
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

State=pd.read_csv('state.csv')
State

```

```python
#2. Numerical variables provide mean, standard deviation, five number summary, histogram
State.describe()
```

```python
State.hist(bins=50)
```

```python
#3. Categorical variables provide a frequency table and bar chart
Freq_State=pd.crosstab(index=State['State'], columns='count')
Freq_Abr=pd.crosstab(index=State['Abbreviation'], columns='count')
Freq_State
Freq_Abr
```

```python
#4. Perform at least 2 pairs of variables for 2 types of bivariate charts
#Plot 1: Murder rate vs. population
import seaborn as sns
sns.set(style="ticks", color_codes=True)

sns.pairplot(State,x_vars=['Population'],y_vars=['Murder.Rate']).set(title="Murder rate vs. Population")

import matplotlib.pyplot as plt
plt.show()
```

```python
#Plot two: Murder rate by state
sns.set(font_scale = 0.5)
sns.barplot(x='Abbreviation',y='Murder.Rate',data=State).set(title="Murder rate by state")
plt.xticks(rotation=45)
plt.show()
```
